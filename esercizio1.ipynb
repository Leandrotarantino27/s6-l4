```python
import pandas as pd
import matplotlib.pyplot as plt
dataframe = pd.read_csv('C:\\Users\\leand\\Desktop\\stockdata.csv')
msft = dataframe['MSFT']
plt.plot(msft, 'b-')
plt.show()
msft_head = dataframe['MSFT'].iloc[:5]
date_head = dataframe['Date'].iloc[:5]
plt.plot(date_head, msft_head, 'ro')
plt.show()
msft_tail = dataframe['MSFT'].iloc[-5:]
date_tail = dataframe['Date'].iloc[-5:]
plt.plot(date_tail, msft_tail, 'g*')
plt.show()
```
