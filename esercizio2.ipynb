```python
import pandas as pd
import matplotlib.pyplot as plt
dataframe = pd.read_csv('C:\\Users\\leand\\Desktop\\stockdata.csv')
aapl = dataframe['AAPL'].iloc[:20]
date = dataframe['Date'].iloc[:20]
plt.plot(date, aapl, 'r--o', markerfacecolor='k')
plt.xlabel('Data')
plt.ylabel('Valore')
plt.title('Azioni Apple')
plt.linewidth=2
plt.xticks([0, 5, 10, 15, 20])
plt.gcf().autofmt_xdate()
plt.tight_layout()
plt.show()
```
